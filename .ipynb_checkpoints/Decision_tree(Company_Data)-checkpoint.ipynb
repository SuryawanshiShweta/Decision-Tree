Robust Scaler Transformation

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_train_robust_scaler = pd.DataFrame(scaler.fit_transform(x_train))
x_train_robust_scaler.head()
```

```python
x_test_robust_scaler = pd.DataFrame(scaler.transform(x_test))
x_test_robust_scaler.head()
```

```python
Robust_model = DecisionTreeClassifier()
```

```python
Robust_model.fit(x_train_robust_scaler,y_train)
```

```python
Robust_model.score(x_train_robust_scaler,y_train)
```

```python
Robust_model.score(x_test_robust_scaler,y_test)
```

```python
#Spliting data into train and test data 
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=3,test_size=0.30)
```

Square Root Transformation

```python
x_train_sr = x_train**(1/2)
x_test_sr = x_test**(1/2)
```

```python
SR_model = DecisionTreeClassifier()
```

```python
SR_model.fit(x_train_sr,y_train)
```

```python
SR_model.score(x_train_sr,y_train)
```

```python
SR_model.score(x_test_sr,y_test)
```

```python
#square transformation
x_train_s=x_train**2
x_test_s=x_test**2
```

```python
#apply this transforemed date in model
S_model  = DecisionTreeClassifier()
S_model.fit(x_train_s,y_train)
```

```python
S_model.score(x_train_s,y_train)
```

```python
S_model.score(x_test_s,y_test)
```

```python
#Creating Dataframe of all model scores
data = pd.DataFrame({'Model No.':['Model','Model1','Model2','Model3','Model4'],
                     'Train_Score':[model.score(x_train,y_train),Min_Max_model.score(x_train_minmax,y_train),Robust_model.score(x_train_robust_scaler,y_train),SR_model.score(x_train_sr,y_train),S_model.score(x_train_s,y_train)],
                     'Test_Score': [model.score(x_test,y_test),Min_Max_model.score(x_test_minmax,y_test),Robust_model.score(x_test_robust_scaler,y_test),SR_model.score(x_test_sr,y_test),S_model.score(x_test_s,y_test)]})
                     
```
