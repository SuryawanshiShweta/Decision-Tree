```python
x = df_new.drop(['Output_Good'],axis=1)
```

Insight :

-  X is an independent variable and we are assigning value to it.
-  we drop Output_Good column because it our output variable 

```python
y = df_new.Output_Good
```

Insight:

-  Y is a dependent variable and we are assigning value to it.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

Insight :

-  With the help of train_test split we divided our dataframe into two parts (train and test)
-  we divided our data into train and test data into 70-30 size means our train data size is 70% of the whole data and test data size is 30% of the whole data
-  we divided our data in two data set because we perform all EDA, visualization, taking some insights from train data, and taken sll insights will apply on test data to predict and to check how good our model is work

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

Insight :

-  Creating model for the prediction
-  here we are using Decision Tree Classifier model because our output value is catogorical value.

```python
model.fit(x_train,y_train)
```

Insight :

-  .fit is use to fit data in model 
-  Here we are provide training data to model for fiting purpose

```python
y_predict = model.predict(x_test)
```

```python
actual_vs_predict = pd.DataFrame({'Actual': y_test,'Prediction': y_predict})
```

```python
actual_vs_predict.head()
```

Insights : 

-  We created the new dataframe of the predicated data and actual data to check our prediction done properly or not.

```python
from sklearn.metrics import classification_report, confusion_matrix
```

Insights :

-  Here confusion matrix contains the number of correct and incorrect predictions of the classification model

-  Classification report is a performance evaluation metric. it is use to show precision, recall and F1 score.

```python
print('Classifiction metrics : \n', classification_report(y_test,y_predict))
```
